##  Building an auto encoder model for Embeddings

```python
def handle_strings(input_str):
    output = str(input_str).lstrip("'").rstrip("'")
    output = ''.join(e for e in output if e.isalnum())
    return output
                     

context_feat_df['categories'] = context_feat_df.categories.apply(lambda x: str(x).lstrip('"[').rstrip(']"').lower().split(', '))
context_feat_df['categories'] = context_feat_df.categories.apply(lambda x: [handle_strings(each) for each in x])
category_list = list(context_feat_df['categories'])
```

```python
category_list
```

```python
vocabulary = set([category for each_list in category_list for category in each_list])
vocabulary = sorted(vocabulary)
print(len(vocabulary))
vocabulary_dict = dict()
for i, word in enumerate(vocabulary):
    vocabulary_dict[word]=i
    
list(vocabulary_dict.items())[:5]
```

```python
def get_one_hot(input_vect):
    output_vect = np.zeros((len(vocabulary),))
    indexes = [vocabulary_dict[each_word] for each_word in input_vect]
    output_vect[indexes]=1
    return output_vect
# test = ['acaibowls','accessories','adultentertainment','afghan']
# get_one_hot(test)
context_feat_df['categories_onehot'] = context_feat_df.categories.apply(lambda x:get_one_hot(x))
```

```python
## build the auto encoder network to train and get embeddings
exclude_columns = ['city', 'state', 'categories', 'is_open', 'latitude', 'longitude','review_count', 'stars','Monday', \
                    'Tuesday','Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday','Alcohol',\
                   'NoiseLevel','RestaurantsAttire','RestaurantsPriceRange2','categories_onehot']
features_select_df = context_feat_df[context_feat_df.columns.difference(exclude_columns)]

context_feat_matrix = np.asarray(features_select_df, dtype=float)
print(context_feat_matrix.shape)
word_one_hot = np.stack(context_feat_df['categories_onehot'])
print(word_one_hot.shape)
```

```python
final_input_ae = np.concatenate([context_feat_matrix,word_one_hot], axis=1)
final_input_ae.shape
```

```python
import warnings
warnings.filterwarnings('ignore')
import keras
print(keras.__version__)
```

```python
from sklearn.model_selection import train_test_split
from keras.layers import Input, Dense
from keras.models import Model
from keras.optimizers import SGD, Adam
import datetime
import os
from keras.callbacks import ModelCheckpoint
```

```python
#train and test split
VALID_SIZE = 0.15
X_train, X_validation = train_test_split(final_input_ae, test_size=VALID_SIZE, random_state=133610)
print("TRAINING DATA")
print('Shape of test sequences: {}'.format(X_train.shape))
print('Shape of validation sequences: {}'.format(X_validation.shape))
```

```python
def generate_samples(x_data, batch_size):
    while True:
        data_size = x_data.shape[0]
        for index in range(0, data_size, batch_size):
            x_temp = x_data[index:index+batch_size]
            yield x_temp, x_temp
```

```python
# next(generate_samples(X_train, 128)).shape
```

```python
encoding_dim = 10
input_dim = X_train.shape[1]
def build_autoencoder():
    input_layer = Input(shape=(input_dim,), name='input_layer')
    encoded = Dense(encoding_dim, activation='relu', name='dense_encoding')(input_layer)
    decoded = Dense(input_dim, activation='sigmoid', name='dense_decoding')(encoded)
    autoencoder = Model(input_layer, decoded)
    enc_model = Model(input_layer, encoded)
    print(autoencoder.summary())
    autoencoder.compile(
      loss='binary_crossentropy',
      optimizer=Adam(lr=0.01),
      metrics=['accuracy']
    )
    return autoencoder, enc_model
```

```python
autoencoder, encmodel = build_autoencoder() 
```

```python
autoencoder.fit(X_train, X_train,
                epochs=50,
                batch_size=128,
                shuffle=True,
                validation_data=(X_validation, X_validation),
               verbose=1)
```

```python
autoencoder.save_weights(BASE_PATH+'autoencoder_model.h5')
```

```python
encoded_values = encmodel.predict(final_input_ae)
```
